```python
from ultralytics import YOLO
import ultralytics
ultralytics.checks()
```

```python
import matplotlib.pyplot as plt
import cv2

model = YOLO('yolov8s.pt')

model.train(data='/content/drive/MyDrive/License Plate Recognition.v8i.yolov8/data.yaml', epochs=70, batch=16, imgsz=640)

model.save('/content/drive/MyDrive/License Plate Recognition.v8i.yolov8/trained_model.pt')
```

```python
model = YOLO('/content/drive/MyDrive/License Plate Recognition.v8i.yolov8/trained_model.pt')
results = model('/content/drive/MyDrive/License Plate Recognition.v8i.yolov8/valid/images/006017783a8beb66_jpg.rf.2a9df3db669e7c82c3dc4046adb0f01a.jpg', show=True)

img = results[0].plot()  # 첫 번째 결과를 시각화
plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
